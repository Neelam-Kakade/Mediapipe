```python
import cv2
import numpy as np
import cvzone
from cvzone.PoseModule import PoseDetector
import math

left_counter = 0
right_counter = 0
left_direction = 0
right_direction = 0

cap = cv2.VideoCapture(r'C:\Users\User\Downloads\LR1.mp4')
pd = PoseDetector(trackCon=0.70, detectionCon=0.70)

def update_counter(counter, direction, angle_interp):
    if angle_interp >= 70:
        if direction == 0:
            counter += 0.5
            direction = 1
    else:
        if direction == 1:
            counter += 0.5
            direction = 0
    return counter, direction

def mouse_callback(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        param['running'] = False

cv2.namedWindow('frame')
cv2.setMouseCallback('frame', mouse_callback, {'running': True})

while True:
    ret, img = cap.read()
    if not ret or not param['running']:
        break

    img = cv2.resize(img, (1000, 500))
    cvzone.putTextRect(img, 'AI Push Up Counter', [345, 30], thickness=2, border=2, scale=2.5)
    pd.findPose(img, draw=0)
    lmlist, bbox = pd.findPosition(img, draw=0, bboxWithHands=0)

    if len(lmlist) != 0:
        left_foot_angle = math.degrees(math.atan2(lmlist[27][2] - lmlist[26][2], lmlist[27][1] - lmlist[26][1]) -
                                        math.atan2(lmlist[24][2] - lmlist[26][2], lmlist[24][1] - lmlist[26][1]))

        right_foot_angle = math.degrees(math.atan2(lmlist[31][2] - lmlist[30][2], lmlist[31][1] - lmlist[30][1]) -
                                         math.atan2(lmlist[28][2] - lmlist[30][2], lmlist[28][1] - lmlist[30][1]))

        left_foot_angle_interp = int(np.interp(left_foot_angle, [-30, 180], [100, 0]))
        right_foot_angle_interp = int(np.interp(right_foot_angle, [-30, 180], [100, 0]))

        left_counter, left_direction = update_counter(left_counter, left_direction, left_foot_angle_interp)
        right_counter, right_direction = update_counter(right_counter, right_direction, right_foot_angle_interp)

        # Display the counters
        cv2.rectangle(img, (0, 0), (120, 120), (255, 0, 0), -1)
        cv2.putText(img, "Left: " + str(int(left_counter)), (20, 70), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 1.6, (0, 0, 255), 7)
        cv2.putText(img, "Right: " + str(int(right_counter)), (20, 120), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 1.6, (0, 0, 255), 7)

        # Display foot movements
        left_val = np.interp(left_foot_angle_interp, [0, 100], [400, 200])
        right_val = np.interp(right_foot_angle_interp, [0, 100], [400, 200])

        cv2.putText(img, 'R', (24, 195), cv2.FONT_HERSHEY_DUPLEX, 1, (255, 0, 255), 7)
        cv2.rectangle(img, (8, 200), (50, 400), (0, 255, 0), 5)
        cv2.rectangle(img, (8, int(right_val)), (50, 400), (255, 0, 0), -1)

        cv2.putText(img, 'L', (962, 195), cv2.FONT_HERSHEY_DUPLEX, 1, (255, 0, 255), 7)
        cv2.rectangle(img, (952, 200), (995, 400), (0, 255, 0), 5)
        cv2.rectangle(img, (952, int(left_val)), (995, 400), (255, 0, 0), -1)

        if left_foot_angle_interp > 70:
            cv2.rectangle(img, (952, int(left_val)), (995, 400), (0, 0, 255), -1)

        if right_foot_angle_interp > 70:
            cv2.rectangle(img, (8, int(left_val)), (50, 400), (0, 0, 255), -1)

    cv2.imshow('frame', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp

# Initialize MediaPipe human pose detection
mp_holistic = mp.solutions.holistic

# Initialize VideoCapture
cap = cv2.VideoCapture(r'C:\Users\User\Downloads\LR1.mp4')  # Replace 'your_video.mp4' with your video file

# Initialize holistic model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Convert the image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        # Make detection
        results = holistic.process(image)
        
        # Draw detected humans on the frame
        if results.pose_landmarks:
            # Draw pose landmarks
            mp.solutions.drawing_utils.draw_landmarks(frame, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)
        
        # Display output
        cv2.imshow('Human Detection', frame)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
